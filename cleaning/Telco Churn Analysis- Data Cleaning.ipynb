### From the above, we can see that the few rows that have the NaN values, have tenure=0. Tenure indicates how many months the customer has been with the company, indicating that they are new and have not yet been charged. Thus, it is safe to remove the rows from the analysis

```python
# dropping the null values, to avoid potential problems in the analysis
data.dropna(inplace= True)
```

# Cleaning

```python
# copying the data secures that the initial data will be intact in case of an error.
data2=data.copy()
# CustomerID provides no information, so we can drop it safely
data2.drop('customerID',axis=1, inplace= True)
# For purposes of readability, we change the SenioCitizen value from 0/1 to No/Yes, to be uniform with the other columns
data2['SeniorCitizen'] = data2['SeniorCitizen'].replace({0: 'No', 1: 'Yes'})
# Churn is our target variable. We change it to boolean in case it needs to be mathematicaly represented in analysis
data2['Churn']= data2['Churn'].replace({'No': False, 'Yes': True})
data2.info()
```

```python
# saving the cleaned data to be used in analysis
data2.to_csv('../data/data_cleaned.csv', index=False)
```
